```python
"""ISIBF score for each city"""

# Function to calculate ISIBF

def calculate_isibf1(num_agencies, neighbors, alpha, threshold):
    isibf_values = {}

    for city in num_agencies:
        total = 0
        for neighbor, distance in neighbors[city].items():
            if distance <= threshold:
                total += num_agencies[neighbor] / alpha**distance
                if city=="Man":
                    print(f"{city} : {neighbor} = {num_agencies[neighbor] / alpha**distance}")
        total+= num_agencies[city] 

        isibf_values[city] = total

    return isibf_values




```

```python

#benin_togo_civ=benin_togo_civ.rename(columns={'adm1_name': 'admin1Name'})


# Calculer les scores ISIBF pour le Bénin
alpha = 1.01
threshold = 400 

#min max normalization
def normalize_scores(scores):
    min_score = min(scores.values())
    max_score = max(scores.values())
    return {city: (score - min_score) / (max_score - min_score) for city, score in scores.items()}

# Calculer les scores ISIBF pour le Bénin
isibf_scores_benin = calculate_isibf1(nb_agencies_benin, neighbors_benin, alpha, threshold)
scores_dep_benin = {department_mapping_benin[city]: score for city, score in isibf_scores_benin.items()}
scores_dep_benin_norm = normalize_scores(scores_dep_benin)


# Calculer les scores ISIBF pour le Togo
isibf_scores_togo = calculate_isibf1(nb_agencies_togo, neighbors_togo, alpha, threshold)
scores_dep_togo = {department_mapping_togo[city]: score for city, score in isibf_scores_togo.items()}
scores_dep_togo_norm = normalize_scores(scores_dep_togo)


# Calculer les scores ISIBF pour la Côte d'Ivoire en utilisant le mapping des départements pour le nombre d'agences
isibf_scores_civ = calculate_isibf1(nb_agencies_civ, neighbors_civ, alpha, threshold)
scores_dep_civ_norm = normalize_scores(isibf_scores_civ)

print(scores_dep_civ_norm)

# Fusionner les scores des deux pays
scores_dep_combined_norm = {**scores_dep_benin_norm, **scores_dep_togo_norm, **scores_dep_civ_norm}

isibf_scores_all= {**scores_dep_benin, **scores_dep_togo, **isibf_scores_civ}
print(isibf_scores_all)
scores_dep_combined_all=normalize_scores(isibf_scores_all)
print(scores_dep_combined_all)

# arrondir les score à 2 chiffres après la virgule
scores_dep_combined_norm = {dep: round(score, 2) for dep, score in scores_dep_combined_norm.items()}
scores_dep_combined_all = {dep: round(score, 2) for dep, score in scores_dep_combined_all.items()}
```

```python
# indicateur de nombre d'agences par région 

# indicateur géographique 
def indicateur_geo(nb_agences, areas):
    ind_scores = {}
    for area in areas:
        ind_scores[area] = nb_agences[area]/areas[area]
    return ind_scores

# indicateur démographique
def indicateur_demo(nb_agences, population, mapping):
    ind_scores = {}
    for dep in nb_agences:  # Itérer sur les départements ou régions
        # Vérifier si le département est dans le mapping
        if dep in mapping:
            region = mapping[dep]  # Trouver la région correspondante dans le mapping
        else:
            region = dep  # Si le département n'a pas besoin de mapping, on l'utilise tel quel
        
        # Calculer l'indicateur démographique
        ind_scores[dep] = (nb_agences[dep] / population[region]) * 100000
    
    return ind_scores


def format_scores(scores, mapping):
    return {mapping.get(city, city): round(score) for city, score in scores.items()}


demo_benin = format_scores(indicateur_demo(nb_agencies_benin, population_benin_over_15, department_mapping_benin), department_mapping_benin)
demo_togo = format_scores(indicateur_demo(nb_agencies_togo, population_togo_over_15, department_mapping_togo), department_mapping_togo)
demo_civ = indicateur_demo(nb_agencies_civ, population_civ_over_15, department_mapping_civ)
#round scores for demo_civ
demo_civ = {dep: round(score) for dep, score in demo_civ.items()}

demo_combined = {**demo_benin, **demo_togo, **demo_civ}

#demo_combined_norm = normalize_scores(demo_combined)

print(demo_combined)

```
