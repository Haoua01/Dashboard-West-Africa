```python
"""ISIBF score for each city"""

# Function to calculate ISIBF

def calculate_isibf1(num_agencies, neighbors, alpha, threshold):
    isibf_values = {}

    for city in num_agencies:
        total = 0
        for neighbor, distance in neighbors[city].items():
            if distance <= threshold:
                total += np.log2(num_agencies[neighbor]+1) / alpha**distance
        total+= np.log2(num_agencies[city] +1)

        isibf_values[city] = total

    return isibf_values

def calculate_isibf0(city_counts, neighbors, alpha, threshold):
    contributions_voisins={}
    contribution_propre={}
    isibf_values = {}
    for city in city_counts:
        total = 0
        for neighbor, distance in neighbors[city].items():
            if distance <= threshold:
                total += np.log2(city_counts[neighbor] + 1) / alpha**distance
            else:
                total += 0
        #print(f"{city_to_region[city]}, contribution des voisins = {total:.2f}, contribution propre = {(city_counts[city] + 1):.2f}")
        contributions_voisins[city] = total
        contribution_propre[city] = np.log2(city_counts[city] + 1)
        total+= np.log2(city_counts[city] + 1)
        isibf_values[city] = total
    return contribution_propre, contributions_voisins, isibf_values

def normalize_scores(scores):
    min_score = min(scores.values())
    max_score = max(scores.values())
    return {city: (score - min_score) / (max_score - min_score) for city, score in scores.items()}


```

```python
# population benin, 2015
population_benin = {
    'Alibori': 867463,
    'Atakora': 772262,
    'Atlantique': 1398229,
    'Borgou': 1212249,
    'Collines': 717477,
    'Couffo':745328,
    'Donga':543130,
    'Littoral': 679012,
    'Mono':497243,
    'Oueme':1100404,
    'Plateau':622372,
    'Zou':851580
}

# population togo, 2022
population_togo = {
    'Kara': 985512,
    'Centrale': 795529,
    'Maritime': 1346615+2188376,
    'Plateaux': 1635946,
    'Savanes_Togo': 1143520
}

#population Cote d'Ivoire, 2021
population_civ = {
    'Abidjan': 6321017,
    'Bas-Sassandra': 2687176,
    'Comoé': 1501336,
    'District du Denguélé': 436015,
    'Gôh-Djiboua': 2088440,
    'Lacs': 1488531,
    'Lagunes': 2042623,
    'Montagnes': 3027023,
    'Sassandra-Marahoué': 2720876,
    'Savanes_CIV': 2159434,
    'Vallée du Bandama': 1964929,
    'Woroba': 1184813,
    'Yamoussoukro': 422072,
    'Zanzan': 1344865
}

# calculate population over 15 years old for each country, knowing that each country has 57% of its population over 15 years old
population_benin_over_15 = {dep: 0.57 * pop for dep, pop in population_benin.items()}
population_togo_over_15 = {dep: 0.57 * pop for dep, pop in population_togo.items()}
population_civ_over_15 = {dep: 0.57 * pop for dep, pop in population_civ.items()}
```

```python
# indicateur de nombre d'agences par région 

# indicateur géographique 
def indicateur_geo(nb_agences, areas):
    ind_scores = {}
    for area in areas:
        ind_scores[area] = nb_agences[area]/areas[area]
    return ind_scores

# indicateur démographique
def indicateur_demo(nb_agences, population, mapping):
    ind_scores = {}
    for dep in nb_agences:  # Itérer sur les départements ou régions
        # Vérifier si le département est dans le mapping
        if dep in mapping:
            region = mapping[dep]  # Trouver la région correspondante dans le mapping
        else:
            region = dep  # Si le département n'a pas besoin de mapping, on l'utilise tel quel
        
        # Calculer l'indicateur démographique
        ind_scores[dep] = (nb_agences[dep] / population[region]) * 100000
    
    return ind_scores


def format_scores(scores, mapping):
    return {mapping.get(city, city): np.ceil(score) for city, score in scores.items()}


demo_benin = format_scores(indicateur_demo(nb_agencies_benin, population_benin_over_15, department_mapping_benin), department_mapping_benin)
demo_togo = format_scores(indicateur_demo(nb_agencies_togo, population_togo_over_15, department_mapping_togo), department_mapping_togo)
demo_civ = indicateur_demo(nb_agencies_civ, population_civ_over_15, department_mapping_civ)
#round scores for demo_civ
demo_civ = {dep: round(score) for dep, score in demo_civ.items()}

demo_combined = {**demo_benin, **demo_togo, **demo_civ}

#demo_combined_norm = normalize_scores(demo_combined)

print(demo_combined)

```
